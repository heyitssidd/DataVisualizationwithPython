# WATER QUALITY TESTING DATA ANALYSIS

Water quality is a crucial aspect of environmental management, and it is essential to measure various physical, chemical, and biological parameters to monitor it effectively. This dataset of 200 rows contains measurements of six critical water quality parameters widely used in water quality monitoring and analysis. The dataset provides a representative snapshot of water quality and can be used for various research, education, and decision-making purposes.

The six parameters measured in this dataset are:

pH: pH measures the acidity or basicity of a liquid on a scale from 0 to 14. Values less than 7 indicate acidic conditions, values greater than 7 suggest primary needs and a value of 7 indicates a neutral state.

Dissolved Oxygen (DO): DO measures the amount of oxygen dissolved in water and is essential for aquatic life. High DO levels are crucial for the survival of fish and other marine organisms.

Temperature: Temperature affects various physical, chemical, and biological processes in water bodies. It is an essential factor that influences the rate of many aquatic processes.

Turbidity (NTU): Turbidity is the measure of relative clarity of a liquid. It is an optical characteristic of water and is a measurement of the amount of light that is scattered by material in the water when a light is shined through the water sample. The higher the intensity of scattered light, the higher the turbidity.

Conductivity (ÂµS/cm): Pure water is not a good conductor of electricity. Ordinary distilled water in equilibrium with carbon dioxide of the air has a conductivity of about 10 x 10-6 W-1*m-1 (20 dS/m). Because the electrical current is transported by the ions in solution, the conductivity increases as the concentration of ions increases.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Lets create a sub - dataframe which has the factors which acts as the pollutants (Turbidity, Conductivity) in decreasing order.....

```python
polluted_water_df = water_quality_df.sort_values(['Turbidity (NTU)', 'Conductivity (ÂµS/cm)' ], ascending=False).groupby('pH').head(50)
polluted_water_df
```

Lets rename the column names for smooth operation in visualization....

```python
polluted_water_df.rename(columns={'Sample ID': 'sample_id', 'Temperature (Â°C)':'Temp','Turbidity (NTU)' : 'Turbidity','Dissolved Oxygen (mg/L)':'Dissolved_oxygen','Conductivity (ÂµS/cm)':'conductivity'   }, inplace=True)
polluted_water_df
```

Lets plot a scatter graph in which we can show all the samples with their different Turbidity Level . . . .

```python
sns.set_style('whitegrid')
sns.relplot(x=polluted_water_df.sample_id, y = polluted_water_df.pH, hue = polluted_water_df.Turbidity);
```

Lets plot a scatter graph in which we can show all the samples with their different Conductivity Level . . . .

```python
sns.set_style('whitegrid')
sns.relplot(x=polluted_water_df.sample_id, y = polluted_water_df.pH, hue = polluted_water_df.conductivity);
```

#### Lets take a sample data from the sub-dataframe for the simplicity purpose . . . 

```python
polluted_sample_df = polluted_water_df.sample(20)
polluted_sample_df
```

Lets plot a graph showing both Turbidity Level and Sample ID for Axes values - Conductivity & Temperature.

We can observe that among these sample data, Sample ID 309 is having the highest Conductivity and Temperature with Turbidity value 3.7


```python
sns.set_style("whitegrid")
sns.relplot(x='Temp', y = 'conductivity', data=polluted_sample_df, hue=polluted_sample_df[['Turbidity', 'sample_id']].apply(tuple, axis=1));
```

Now we will plot a graph showing different conductivity level for every Sample ID. 
We can observe that as Sample ID is increasing, Conductivity is also increasing.

```python
sns.scatterplot(x=polluted_water_df.sample_id, y=polluted_water_df.conductivity);
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

Lets plot a BarPlot showing variations for Dissolved Oxygen for every Sample ID. . . 

```python
sns.barplot(y=polluted_water_df.sample_id, x=polluted_water_df.Dissolved_oxygen)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```
